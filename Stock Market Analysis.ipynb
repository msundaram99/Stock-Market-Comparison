```python
import pandas as pd
import yfinance as yf
import plotly.io as pio
import plotly.graph_objects as go
pio.templates.default = "plotly_white"
```

```python
# Define the tickers for Apple and Google
apple_ticker = 'AAPL'
google_ticker = 'GOOGL'

# Define the date range for the last quarter
start_date = '2023-07-01'
end_date = '2023-09-30'

# Fetch historical stock price data using yfinance
apple_data = yf.download(apple_ticker, start=start_date, end=end_date)
google_data = yf.download(google_ticker, start=start_date, end=end_date)
```

```python
print(google_data.head(10))
```

```python
# Calculate daily returns
apple_data['Daily_Return'] = apple_data['Close'].pct_change()
google_data['Daily_Return'] = google_data['Close'].pct_change()
```

```python
print(apple_data.head(10))
```

```python
import plotly.graph_objects as go

# Create the figure
fig = go.Figure()

# Add the Apple daily return trace
fig.add_trace(go.Scatter(x=apple_data.index, y=apple_data['Daily_Return'], mode='lines', name='Apple'))

# Add the Google daily return trace
fig.add_trace(go.Scatter(x=google_data.index, y=google_data['Daily_Return'], mode='lines', name='Google'))

# Update the layout
fig.update_layout(title='Daily Returns of Apple and Google',
                  xaxis_title='Date',
                  yaxis_title='Daily Return',
                  legend=dict(x=0.08, y=0.95))

# Show the plot
fig.show()
```

```python
# prompt: # Calculate cumulative returns for the last quarter

# Calculate cumulative returns
apple_data['Cumulative_Return'] = (1 + apple_data['Daily_Return']).cumprod()
google_data['Cumulative_Return'] = (1 + google_data['Daily_Return']).cumprod()

# Create the figure for cumulative returns
fig_cumulative = go.Figure()

# Add the Apple cumulative return trace
fig_cumulative.add_trace(go.Scatter(x=apple_data.index, y=apple_data['Cumulative_Return'], mode='lines', name='Apple'))

# Add the Google cumulative return trace
fig_cumulative.add_trace(go.Scatter(x=google_data.index, y=google_data['Cumulative_Return'], mode='lines', name='Google'))

# Update the layout for cumulative returns
fig_cumulative.update_layout(title='Cumulative Returns of Apple and Google',
                         xaxis_title='Date',
                         yaxis_title='Cumulative Return',
                         legend=dict(x=0.08, y=0.95))

# Show the cumulative returns plot
fig_cumulative.show()
```

```python
# prompt: Calculate historical volatility (standard deviation of daily returns)

# Calculate historical volatility (standard deviation of daily returns)
apple_volatility = apple_data['Daily_Return'].std()
google_volatility = google_data['Daily_Return'].std()

print(f"Apple Volatility: {apple_volatility}")
print(f"Google Volatility: {google_volatility}")
```

```python
# Create the figure for volatility comparison
fig_volatility = go.Figure()

# Add bars for Apple and Google volatility
fig_volatility.add_trace(go.Bar(x=['Apple'], y=[apple_volatility], name='Apple'))
fig_volatility.add_trace(go.Bar(x=['Google'], y=[google_volatility], name='Google'))

# Update the layout for volatility comparison
fig_volatility.update_layout(title='Volatility Comparison of Apple and Google',
                             xaxis_title='Stock',
                             yaxis_title='Volatility (Standard Deviation)',
                             bargap=0.5) # Remove the legend

# Show the volatility plot
fig_volatility.show()
```

```python
market_data = yf.download('^GSPC', start=start_date, end=end_date)  # S&P 500 index as the
market_data.head(10)
```

```python
# Calculate daily returns for both stocks and the market
market_data['Daily_Return'] = market_data['Close'].pct_change()
market_data
```

```python
# Calculate Beta for Apple and Google
cov_apple = apple_data['Daily_Return'].cov(market_data['Daily_Return'])
var_market = market_data['Daily_Return'].var()

beta_apple = cov_apple / var_market

cov_google = google_data['Daily_Return'].cov(market_data['Daily_Return'])
beta_google = cov_google / var_market
```

```python
# Compare Beta values
if beta_apple > beta_google:
    conclusion = "Apple is more volatile (higher Beta) compared to Google."
else:
    conclusion = "Google is more volatile (higher Beta) compared to Apple."

# Print the conclusion
print("Beta for Apple:", beta_apple)
print("Beta for Google:", beta_google)
print(conclusion)
```
